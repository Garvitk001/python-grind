# Default parameters - agar value na do toh default use hogi

```python
def calculate_discount(price , discount=10):
    final_price = price - (price *discount/100)
    return final_price



print(f"Default Discount :{calculate_discount(300)}  \nCustom_Discount : {calculate_discount(300,20)}")

```

```python
def create_student_profile(name, age, branch="AI & DS", cgpa=0.0):
    return {
        "name": name,
        "age": age,
        "branch": branch,
        "cgpa": cgpa
    }
```

```python
student1 = create_student_profile("Rahul", 21)
print(student1)


student2 = create_student_profile("Garvit", 22, "AI & DS", 8.3)
print(student2)
```

```python

def add_numbers(a, b):
    return a + b

print(add_numbers(2, 3))
```

```python
from numpy import add


def add_numbers(*args):
        return sum(args)

print(add_numbers(2,4))
print(add_numbers(2,5,8))
```

```python
def calculate_marks(*marks):
    return sum(marks)

print (calculate_marks(76,89,86,90))
```

```python
# ============================================
# **kwargs - Keyword Arguments
# ============================================

# *args = bina naam ke values
# **kwargs = naam ke saath values (key=value)
```

```python
def create_profile(**kwargs):
    print(kwargs)
```

```python
create_profile(name= "Garvit",age = 22 , city="Jaipur")
```

```python
def add_employees(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")
```

```python
add_employees(name="Rahul", dept="ML", salary=45000)
print()
add_employees(name="Priya", dept="DS", salary=62000, experience=3)
```

```python

def student_report(*args, **kwargs):
    total= sum(args)
    avg = total/ len(args)
    name = kwargs["name"]
    branch = kwargs["branch"]
    print(f"{name} ({branch})  - Total Marks : {total}  - Average = {avg}")
```

```python
student_report(85, 92, 78, 90, 88, name="Garvit Khuteta", branch="AI & DS")
student_report(55, 96, 48, 90, 84, name="Rahul gupta", branch="CS")
```
