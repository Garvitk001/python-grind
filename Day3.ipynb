PASSWORD VALIDATOR

```python

from numpy import append


def validate_password(password):
     has_upper = any(char.isupper() for char in password)
     has_digit = any(char.isdigit() for char in password)
     has_special = any(char in "@!#%&$" for char in password) 
     has_length= len(password)>8

     score = has_upper + has_digit + has_length + has_special

     reason= []

     if not has_length  : reason.append("too short ")
     if not has_special: reason.append("No special character")
     if not has_upper: reason.append("no capital letter")
     if not has_digit : reason.append("No number")

     if score == 4:
          return "Strong Password!"
     elif score == 3:
          return f"Medium Passsword!....Missing : {','.join(reason)}"
     else:
          return f"weak password!......Missing : {','.join(reason)}"
```

```python
print(validate_password("Garvit@123"))
print(validate_password("garvit"))
print(validate_password("Garvit123"))  
```

```python
class Student:
    def __init__(self,name,age,cgpa):
        self.name = name
        self.age = age
        self.cgpa = cgpa
    
    def introduce(self):
        print(f" Hi i am {self.name} , my age is {self.age} and my cgpa is {self.cgpa}")
    

    def get_grade(self):
        if self.cgpa >=9.0:
             return "Outstanding"
        elif self.cgpa >= 8.0:
            return "Excellent"
        elif self.cgpa >= 7.0:
            return "Good"
        else:
            return "Average"


garvit = Student("Garvit Khuteta", 22, 8.3)
rahul  = Student("Rahul Sharma",   21, 7.5)
priya  = Student("Priya Singh",    22, 9.2)


# Methods call karo
garvit.introduce()
rahul.introduce()
priya.introduce()

print(f"\n{garvit.name} ka grade : {garvit.get_grade()}")

```

```python
class BankAccount:
    def __init__ (self, account_holder , account_number , initial_balance = 0):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = initial_balance
        self.transactions = []

    def deposit(self,amount):
        if amount<=0:
            return " INVALID AMOUNT!"
        else:
            self.balance += amount
            self.transactions.append(f"+ ${amount:,}")
            return f"${amount:,} deposited ...Balance : ${self.balance:,}"
        
    def withdraw(self , amount):
        if amount <=0:
            return " INVALID AMOUNT"
        elif amount > self.balance:
            return "UNSUFFICIENT BALANCE"
        self.balance-= amount
        self.transactions.append(f"-${amount:,}")
        return f"${amount:,} withdraw.....Balance : ${self.balance:,}"
    
    def check_balance(self):
        return f"Balance : ${self.balance:,}"
    
    def get_statement(self):
        print(f"\n{'='*40}")
        print(f"  Account: {self.account_holder}")
        print(f"  Acc No : {self.account_number}")
        print(f"{'='*40}")
        for t in self.transactions:
            print(f"  {t}")
        print(f"{'─'*40}")
        print(f"  Final Balance: ₹{self.balance:,}")
        print(f"{'='*40}")
```

```python
garvit_acc = BankAccount("Garvit Khuteta", "SBI001", 10000)


print(garvit_acc.deposit(5000))
print(garvit_acc.deposit(3000))
print(garvit_acc.withdraw(2000))
print(garvit_acc.withdraw(50000)) 
print(garvit_acc.withdraw(-500))   

# Statement 
garvit_acc.get_statement()
```

```python
class Product:
    def __init__(self , name, price, stock,category):
        self.name = name
        self.price = price
        self.stock = stock
        self.category = category
    

    def get_info(self):
        return f"=====Product Info=====\n Name : {self.name}\n Category : {self.category}\n Price : ${self.price}\n Stock : {self.stock}"
    
    def apply_discount(self,discount):
        discount_price = self.price - (self.price * (discount/100))
        return f" After applying {discount}% on {self.name}  new price is ${discount_price} "
    

    def add_stock(self,update):
        if update <=0:
            return "Invalid Stock!"
        self.stock += update
        return f" Total stock  of {self.name}  is {self.stock}" 
    
    def sell_stock(self,quantity):
        if quantity > self.stock:
            return f"OUT OF STOCK!!!"
        self.stock -= quantity
        return f"Totsl sold {quantity} of {self.name} and remaining are {self.stock}" 
    
    
```

```python
# Object banao
iphone = Product("iPhone 15", 79999, 50, "Electronics")


print(iphone.get_info())
print(iphone.apply_discount(10))
print(iphone.add_stock(20))
print(iphone.sell_stock(15))
print(iphone.sell_stock(200))  # Out of stock case
```

```python
# Inheritance = ek class doosri class ki properties le leti hai
# Real life: 
# Animal class hai — Dog aur Cat us se inherit karte hain
# Animal ke saare kaam Dog aur Cat kar sakte hain + apne extra kaam
```

```python
class Animal:
    def __init__(self,name):
        self.name = name
    
    def eat(self):
        return f"{self.name} is eating "
    
class Dog(Animal):
    def bark(self):
        return f"{self.name} is barking"
    
class Cat(Animal):
    def meow(self):
        return f" {self.name} is meowing "


# Test karo
dog = Dog("Tommy")
cat = Cat("Kitty")
print(dog.eat())    # Animal ka method — Dog pe kaam karega!
print(dog.bark())   # Dog ka apna method
print(cat.eat())    # Animal ka method — Cat pe bhi kaam karega!
print(cat.meow())   # Cat ka apna method
```

# LIBRARY MANAGEMENT SYSTEM

```python
class LibraryItem:
    def __init__(self, title, author, year, copies):
        self.title = title
        self.author = author
        self.year = year
        self.copies=copies
        self.issued=0
    
    def get_info(self):
        print(f"\n{'='*45}")
        print(f"  Title     : {self.title}")
        print(f"  Author    : {self.author}")
        print(f"  Year      : {self.year}")
        print(f"  Available : {self.copies-self.issued}/{self.copies}")
        print(f"{'='*45}")

    def issue_item(self,member_name):
        if self.issued >=self.copies:
            return f"{self.title} is not available!"
        self.issued +=1
        return f" {self.title} issue to {member_name}"

    def return_item(self,member_name):
        if self.issued <=0:
            return f"No copy is isssued"       
        self.issued -= 1
        return f"{self.title} return by {member_name}"
    
#super() sai apn parent ke function call krte h

class Book(LibraryItem):
    def __init__(self,title,author,year,copies,genre,pages):
        super().__init__(title,author,year,copies)
        self.genre = genre
        self.pages=pages

    def get_info(self):
        super().get_info()   
        print(f"  Genre     : {self.genre}")
        print(f"  Pages     : {self.pages}")
        print(f"{'='*45}")

class Magazine(LibraryItem):
    def __init__(self, title, author, year, copies, issue_number):
        super().__init__(title, author, year, copies)
        self.issue_number = issue_number
    
    def get_info(self):
        super().get_info()
        print(f"  Issue No  : {self.issue_number}")
        print(f"{'='*45}")
```

```python
python_book = Book("Python Crash Course", "Eric Matthes", 2019, 3, "Programming", 544)
ds_magazine = Magazine("Data Science Weekly", "John Doe", 2024, 5, 142)

python_book.get_info()
ds_magazine.get_info()

# Issue karo
print(python_book.issue_item("Garvit Khuteta"))
print(python_book.issue_item("Rahul Sharma"))
print(python_book.issue_item("Priya Singh"))
print(python_book.issue_item("Amit Verma"))   # ye fail hona chahiye!

# Return karo
print(python_book.return_item("Garvit Khuteta"))

# Magazine issue karo
print(ds_magazine.issue_item("Garvit Khuteta"))
```
