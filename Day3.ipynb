# Default parameters - agar value na do toh default use hogi

```python
def calculate_discount(price , discount=10):
    final_price = price - (price *discount/100)
    return final_price



print(f"Default Discount :{calculate_discount(300)}  \nCustom_Discount : {calculate_discount(300,20)}")

```

```python
def create_student_profile(name, age, branch="AI & DS", cgpa=0.0):
    return {
        "name": name,
        "age": age,
        "branch": branch,
        "cgpa": cgpa
    }
```

```python
# ============================================
# **kwargs - Keyword Arguments
# ============================================

# *args = bina naam ke values
# **kwargs = naam ke saath values (key=value)
```

```python
def create_profile(**kwargs):
    print(kwargs)
```

```python
create_profile(name= "Garvit",age = 22 , city="Jaipur")
```

```python
def add_employees(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")
```

```python
add_employees(name="Rahul", dept="ML", salary=45000)
print()
add_employees(name="Priya", dept="DS", salary=62000, experience=3)
```

```python

def student_report(*args, **kwargs):
    total= sum(args)
    avg = total/ len(args)
    name = kwargs["name"]
    branch = kwargs["branch"]
    print(f"{name} ({branch})  - Total Marks : {total}  - Average = {avg}")
```

```python
student_report(85, 92, 78, 90, 88, name="Garvit Khuteta", branch="AI & DS")
student_report(55, 96, 48, 90, 84, name="Rahul gupta", branch="CS")
```

PASSWORD VALIDATOR

```python

from numpy import append


def validate_password(password):
     has_upper = any(char.isupper() for char in password)
     has_digit = any(char.isdigit() for char in password)
     has_special = any(char in "@!#%&$" for char in password) 
     has_length= len(password)>8

     score = has_upper + has_digit + has_length + has_special

     reason= []

     if not has_length  : reason.append("too short ")
     if not has_special: reason.append("No special character")
     if not has_upper: reason.append("no capital letter")
     if not has_digit : reason.append("No number")

     if score == 4:
          return "Strong Password!"
     elif score == 3:
          return f"Medium Passsword!....Missing : {','.join(reason)}"
     else:
          return f"weak password!......Missing : {','.join(reason)}"
```

```python
print(validate_password("Garvit@123"))
print(validate_password("garvit"))
print(validate_password("Garvit123"))  
```
