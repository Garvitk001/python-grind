#LIST OPERATIONS - Sabse zyada use hone wale

```python
products =["Laptop", "Phone","Tablet"]

products.append("Earbuds")                     #last mai add
products.insert(1,"Charger")                  #particular jagah par add
print(f"After add   : {products}")
```

```python
products.remove("Charger")                 #value se remove
products.pop()                              # last item remove
products.pop(0)                            # particular location se remover
print(products)
```

```python
numbers = [45,12,89,3,67,23]
numbers.sort()                           #assending
print(numbers)
numbers.reverse()
print(numbers)
```

```python
print(f"Phone Index :  {products.index("Phone")}")
print(f"Phone exist :  {'Phone' in products }")
print(f"Count       :  {products.count('Phone')}")
```

# LIST COMPREHENSION

```python
prices = [45000, 18000, 22000, 2500, 1500]
```

```python
discounted=[]
for price in  prices:
    discounted.append(price * 0.9)            #normal
print(f"{discounted}")
```

```python
discounted= [price * 0.9 for price in prices]             #one line
print(discounted)
```

```python
expensive = [price for price in prices if price>10000]          #filter
print(expensive)
```

```python
# String list pe bhi
customers = ["garvit", "rahul", "priya", "amit"]
upper_name= [ name.upper() for name in customers]
print(f"{upper_name}")
```

```python
marks = [85, 45, 92, 38, 78, 55, 91]
passed_marks= [ mark for mark in marks if mark >=50]
failed_marks= [ mark for mark in marks if mark<50]

print(f" Passed Mark  : {passed_marks}\n Failed Mark   : {failed_marks}")
```

```python
employees = [
    {"name": "Garvit",  "salary": 25000, "dept": "DS"},
    {"name": "Rahul",   "salary": 45000, "dept": "ML"},
    {"name": "Priya",   "salary": 62000, "dept": "DS"},
    {"name": "Amit",    "salary": 38000, "dept": "ML"},
    {"name": "Deepak",  "salary": 71000, "dept": "DS"},
]
```

```python
#Sirf DS department ke employees ke naam
ds_list = [ emp["name"] for emp in employees if emp["dept"] == "DS"]
print(f"DS Employees   : {ds_list}")
```

```python
#Sirf jinki salary 40000 se zyada hai unke naam
high_salary= [ emp["name"] for emp in employees  if emp["salary"] > 40000]
print(f"High Salary  : {high_salary}")
```

```python
#Har employee ki salary 10% badhao — new list

salary_inc = [ emp["salary"] * 1.1 for emp in employees]
print(f"Salary +10%  : {salary_inc}")
```

```python
#Saari salaries uppercase mein nahi — naam uppercase mein nikalo

upper_name= [emp["name"].upper() for emp in employees]
print(f" Upper Name  : {upper_name}")
     
```
