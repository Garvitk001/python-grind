```python
# ============================================
# IPL PLAYER ANALYSER
# Real use: Sports analytics companies
# ============================================

players = [
    {"name": "Virat Kohli",     "team": "RCB", "runs": 6634, "matches": 223, "avg": 36.2, "role": "Batsman"},
    {"name": "Rohit Sharma",    "team": "MI",  "runs": 6211, "matches": 213, "avg": 31.1, "role": "Batsman"},
    {"name": "MS Dhoni",        "team": "CSK", "runs": 5082, "matches": 234, "avg": 38.1, "role": "Wicketkeeper"},
    {"name": "Jasprit Bumrah",  "team": "MI",  "runs": 56,   "matches": 120, "avg": 5.6,  "role": "Bowler"},
    {"name": "Rashid Khan",     "team": "GT",  "runs": 587,  "matches": 95,  "avg": 14.2, "role": "Bowler"},
    {"name": "Suryakumar Yadav","team": "MI",  "runs": 2644, "matches": 95,  "avg": 31.5, "role": "Batsman"},
    {"name": "Hardik Pandya",   "team": "GT",  "runs": 2309, "matches": 107, "avg": 28.4, "role": "All-rounder"},
    {"name": "KL Rahul",        "team": "LSG", "runs": 4163, "matches": 118, "avg": 46.8, "role": "Batsman"},
    {"name": "Ravindra Jadeja", "team": "CSK", "runs": 2386, "matches": 226, "avg": 26.5, "role": "All-rounder"},
    {"name": "Yuzvendra Chahal","team": "RR",  "runs": 81,   "matches": 145, "avg": 6.2,  "role": "Bowler"},
]
```

```python
# FUNCTION 1
# Naam: get_top_scorer
# Sabse zyada runs wala player return kare (dictionary)

def get_top_scorer():
    top_scorer = ""
    top_runs = 0
    
    for player in players:
        if player["runs"] > top_runs:
            top_runs = player["runs"]
            top_scorer = player["name"]
    
    return {"name": top_scorer, "runs": top_runs}   # dictionary return karo

# Test karo
result = get_top_scorer()
print(result)
```

```python
# FUNCTION 2  
# Naam: get_players_by_team
# Input: team name (string)
# Output: us team ke saare players ki list


def get_players_by_team(team_name):
    result = []
    
    for player in players:
        if player["team"] == team_name:
            result.append(player["name"])
    
    return result

print(get_players_by_team("MI"))
print(get_players_by_team("CSK"))
        
```

```python
# FUNCTION 3
# Naam: get_players_by_role
# Input: role (string) - "Batsman", "Bowler", etc
# Output: us role ke players ki list (List Comprehension use karo)


def get_players_by_role(category):
    result = []

    for player in players:
        if player["role"] == category:
            result.append(player["name"])
    return result

print(f"Batman are {get_players_by_role("Batsman")} \nBowler are {get_players_by_role("Bowler")} ")
```

```python
# FUNCTION 4
# Naam: calculate_strike_rate_category
# Input: avg (float)
# Output: "Excellent" (avg>35), "Good" (avg>25), "Average" 

def calculate_strike_rate_category(avg):
    if avg > 35:
        return "Excellent"
    elif avg > 25:
        return "Good"
    else:
        return "Average"

print(calculate_strike_rate_category(38.1))  
print(calculate_strike_rate_category(28.4))  
print(calculate_strike_rate_category(5.6))   
```

```python
# FUNCTION 5
# Naam: get_team_total_runs
# Input: team name
# Output: us team ke saare players ke total runs

def get_team_total_run(team_name):
    total_run = 0
    for player in players:
        if player["team"] == team_name:
            total_run += player["runs"]
    return total_run

print(get_team_total_run("MI"))
```

```python
# ============================================
# FINAL IPL REPORT
# ============================================

print("=" * 50)
print("       IPL PLAYER ANALYSIS REPORT")
print("=" * 50)

# 1. Top Scorer
top_scorer = get_top_scorer()
print(f" Top Scorer    : {top_scorer['name']} ({top_scorer['runs']} runs)")

# 2. Team wise players
print(f"\n MI Players    : {get_players_by_team('MI')}")
print(f" CSK Players   : {get_players_by_team('CSK')}")

# 3. Role wise
print(f"\nBatsmen       : {get_players_by_role('Batsman')}")
print(f"Bowlers       : {get_players_by_role('Bowler')}")

# 4. Strike rate 
print(f"\n Performance Categories:")
for player in players:
    category = calculate_strike_rate_category(player["avg"])
    print(f"   {player['name']:<20} : {category}")

# 5.   Team   total runs
print(f"\n MI Total Runs  : {get_team_total_run('MI')}")
print(f" CSK Total Runs : {get_team_total_run('CSK')}")

print("=" * 50)
```
