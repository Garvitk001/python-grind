```python
# ============================================
# CHALLENGE - Student Database
# ============================================

students = {
    "S001": {"name": "Garvit Khuteta", "marks": [85, 92, 78, 90, 88], "city": "Jaipur"},
    "S002": {"name": "Rahul Sharma",   "marks": [65, 70, 58, 72, 68], "city": "Delhi"},
    "S003": {"name": "Priya Singh",    "marks": [92, 95, 98, 91, 96], "city": "Mumbai"},
    "S004": {"name": "Amit Verma",     "marks": [45, 52, 48, 55, 50], "city": "Jaipur"},
    "S005": {"name": "Deepak Gupta",   "marks": [75, 80, 72, 85, 78], "city": "Delhi"},
}
```

```python
from turtle import st


topper_name = ""
topper_avg  = 0
pass_students = 0

for roll , student in students.items():
    marks = student["marks"]
    avg = sum(marks) / len(marks)   

    if avg >= 50:
        pass_students+= 1
    
    if avg > topper_avg:
        topper_avg = avg
        topper_name = student["name"]

    print(f" {roll}   |  {student['name']:<18}  |  Avg: {avg}")
```

```python
jaipur_students = [student["name"] 
                   for roll, student in students.items() 
                   if student["city"] == "Jaipur"]
print(f"Jaipur students : {jaipur_students}")
```

```python

print(f"Topper : {topper_name} ({topper_avg}%)")
```
